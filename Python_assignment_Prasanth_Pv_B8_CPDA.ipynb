```python
import pandas as pd

columns = [
"year","make","model","trim","body","transmission","vin","state",
"condition","odometer","color","interior","seller",
"mmr","sellingprice","saledate"
]

df = pd.read_csv("car_prices.csv", names=columns, skiprows=1)

print(df.head())
print(df.columns)
print("sellingprice" in df.columns)
```

```python
print("Average Selling Price:", df['sellingprice'].mean())
print("Maximum Selling Price:", df['sellingprice'].max())
print("Minimum Selling Price:", df['sellingprice'].min())
```

2.12 Best Value Brands (Top 20% Condition)
Find top 20% condition threshold

```python
threshold = df['condition'].quantile(0.8)
```

Filter and calculate cheapest brands

```python
df[df['condition'] >= threshold].groupby('make')['sellingprice'].mean().sort_values()
```
